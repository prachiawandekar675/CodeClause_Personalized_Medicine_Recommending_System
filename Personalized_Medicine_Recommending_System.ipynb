```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

#stop_words='english': This parameter specifies that common English stop words should be removed from the text data. Stop words are words that are frequently used in a language but typically don't contribute much meaning to the text, such as "and," "the," "is," etc. By setting stop_words to 'english', the CountVectorizer will automatically filter out these common stop words during tokenization.
#max_features=5000: This parameter specifies the maximum number of features (words) to consider when building the vocabulary. The vocabulary contains unique words from the input text data, and setting max_features limits the vocabulary size to the specified number of most frequent words. This can help control the dimensionality of the resulting feature matrix.
cv = CountVectorizer(stop_words='english',max_features=5000)
```

Define the stem() function: This function takes a string of text as input and performs stemming on each word in the text. It splits the input text into words using whitespace as a delimiter, then applies stemming to each word using the PorterStemmer. The stemmed words are collected in the list y.

Return the stemmed text: The stemmed words are joined back together using spaces to form a single string, which is then returned by the function.


```python
def stem(text):
  y = []

  for i in text.split():
    y.append(ps.stem(i))

  return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

cv.fit_transform(new_df['tags']): This fits the CountVectorizer object (cv) to the text data in the 'tags' column of the DataFrame new_df and transforms it into a sparse matrix using the bag-of-words representation.

.toarray(): This converts the sparse matrix into a dense NumPy array. The toarray() method is used to retrieve the dense representation of the sparse matrix.

.shape: This returns the shape of the resulting NumPy array, which represents the dimensions of the matrix

```python
cv.fit_transform(new_df['tags']).toarray().shape
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

The cosine_similarity() function is a commonly used metric to calculate the cosine similarity between pairs of vectors. In the context of natural language processing (NLP), it's often used to measure the similarity between document vectors, where each document is represented by a vector in a high-dimensional space (such as a bag-of-words representation).

```python
cosine_similarity(vectors)
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[1]
```

medicine_index: This line finds the index of the input medicine name in the DataFrame new_df. It assumes that the DataFrame contains a column named 'Drug_Name' that stores medicine names.

distances: This line retrieves the distances (similarities) between the input medicine and all other medicines in the similarity matrix. The similarity matrix is indexed using medicine_index.

medicines_list: This line creates a list of tuples where each tuple contains the index of a medicine and its corresponding similarity value. The list is sorted in descending order of similarity (reverse=True) using the key parameter.

The loop iterates through the top 5 similar medicines (excluding the input medicine itself) and prints their names.


```python
def recommend(medicine):
    medicine_index = new_df[new_df['Drug_Name'] == medicine].index[0]
    distances = similarity[medicine_index]
    medicines_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

    for i in medicines_list:
        print(new_df.iloc[i[0]].Drug_Name)
```

```python
recommend("Paracetamol 125mg Syrup 60mlParacetamol 500mg Tablet 10'S")
```
